```python
#Visualizing the target column
sns.countplot(x='Attrition', data=df, palette='Set3')
plt.title('Employee Attrition Distribution')
plt.show()
```

## 1.2 Univariate Analysis

```python
#Checking the Age column
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(x='Age', data=df, bins=30, color='salmon')
plt.title('Age distribution')

plt.subplot(1,2,2)
sns.histplot(x='MonthlyIncome', data=df, bins=30, color='skyblue')
plt.title('Income distribution')
plt.show()
```

```python
#Checking the Job Satisfaction and Hike Percentage column
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(x='JobSatisfaction', data=df, bins=30, color='pink')
plt.title('Job Satisfaction distribution')

plt.subplot(1,2,2)
sns.histplot(x='PercentSalaryHike', data=df, bins=30, color='yellow')
plt.title('Hike Percentage distribution')
plt.show()
```

### 2. Correlation Check for Numerical features(Heat Map)

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(15,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Dropping the redundant features
df_clean = df.drop(columns=['DailyRate', 'EmployeeCount', 'HourlyRate', 'MonthlyRate', 'StandardHours', 'JobLevel', 'YearsAtCompany', 'Over18', 'TotalWorkingYears', 'EmployeeNumber'])
```
